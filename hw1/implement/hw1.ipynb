# 載入套件

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import os
```

# 影像處理

```python
def show_img(image_path, kernel):
    img = cv2.imread(image_path)
    dst = cv2.filter2D(img, -1, kernel)
    stack = cv2.cvtColor(dst, cv2.COLOR_BGR2RGB)
    plt.imshow(stack)
```

```python
kernel_sharp = np.array((
        [0,-1,0],
        [-1,5,-1],
        [0,-1,0]), dtype="float32")
```

```python
kernel_emboss = np.array((
        [-2,-1,0],
        [-1,1,1],
        [0,1,2]), dtype="float32")
```

```python
kernel_laplacian = np.array((
        [0,1,0],
        [1,-4,1],
        [0,1,0]), dtype="float32")
```

# 顯示圖片

## 原圖片

```python
Image.open('lion.jpg')
```

```python
try:
    os.mkdir("image/")
except:
    pass
```

## sharpen

```python
show_img('lion.jpg', kernel_sharp)
plt.title("sharpen")
plt.axis('off')
plt.savefig("image/sharp.png")
```

## emboss

```python
show_img('lion.jpg', kernel_emboss)
plt.title('emboss')
plt.axis('off')
plt.savefig('image/emboss.png')
```

## laplacian

```python
show_img('lion.jpg', kernel_laplacian)
plt.title('laplacian operator')
plt.axis('off')
plt.savefig('image/laplacian.png')
```
